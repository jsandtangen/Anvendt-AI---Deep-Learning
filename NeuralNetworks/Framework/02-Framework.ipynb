```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Generer datasett
n_samples = 100
features, labels = make_classification(
    n_samples=n_samples,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    flip_y=0.2,
    random_state=42
)

features = features.astype(np.float32)
labels = labels.astype(np.int32)
```

```python
# Bruk sklearn sin egen train_test_split i stedet for np.split
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=67
)
```

```python
def plot_dataset(ax, title, X, y):
    """Plotter et datasett på en gitt matplotlib-akse (ax)."""
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', 
                          s=100, alpha=0.6, edgecolors='k')
    ax.set_title(title, fontsize=13)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    return scatter

# Vis train og test side ved side i samme figur
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plot_dataset(axes[0], 'Treningsdata', X_train, y_train)
plot_dataset(axes[1], 'Testdata', X_test, y_test)
fig.suptitle('Syntetisk klassifikasjonsdatasett', fontsize=16)
plt.tight_layout()
plt.show()
```
